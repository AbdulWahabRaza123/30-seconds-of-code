```python
#let's remove the all stop words from contents column
#Now making preprocessing function
import spacy
nlp=spacy.load("en_core_web_sm")
def preprocess(text):
  doc=nlp(text)
  #removing stop words and punc marks
  non_stop_words=[token.text for token in doc if not token.is_stop and not token.is_punct]
# to get strings in return
  return " ".join(non_stop_words)
```

```python
#creating new column
df['new_contents']=df['contents'].apply(preprocess)
df.head()
```
